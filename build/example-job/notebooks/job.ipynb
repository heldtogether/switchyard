# Sample artefact notebook

This notebook demonstrates writing outputs into `/outputs`.

```python
output_dir = '/outputs'
```

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

out = Path(output_dir)
out.mkdir(parents=True, exist_ok=True)

df = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'value': [10, 18, 14, 25],
})

csv_path = out / 'data.csv'
xlsx_path = out / 'data.xlsx'
png_path = out / 'chart.png'
txt_path = out / 'summary.txt'

df.to_csv(csv_path, index=False)
df.to_excel(xlsx_path, index=False)

plt.figure(figsize=(6, 4))
plt.plot(df['month'], df['value'], marker='o')
plt.title('Sample output chart')
plt.xlabel('Month')
plt.ylabel('Value')
plt.tight_layout()
plt.savefig(png_path, dpi=150)
plt.close()

txt_path.write_text('Notebook ran successfully and wrote artefacts to /outputs\n')

print(f'Wrote: {csv_path}')
print(f'Wrote: {xlsx_path}')
print(f'Wrote: {png_path}')
print(f'Wrote: {txt_path}')
```
